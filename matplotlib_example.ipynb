```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
	return datetime.strptime(x, '%Y-%m-%d')

series = read_csv('QQQ.csv', header=0, parse_dates=['Date'], index_col=0, squeeze=True, date_parser=parser)
print(series.head())
```

```python
series['Close'].plot()
pyplot.show()
```

```python
series['Volume'].plot()
pyplot.show()
```
